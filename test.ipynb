```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Récupérer les données
url = "https://data.airpl.org/api/v1/mesure/journaliere/"
params = {
    "code_configuration_de_mesure__code_point_de_prelevement__code_polluant": 24,
    "date_heure_tu__range": "2024-3-1,2024-3-31",
    "code_configuration_de_mesure__code_point_de_prelevement__code_station__code_commune__code_departement__in": "44,49,53,72,85",
    "export": "json"
}

response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data['results'])  # 'results' peut changer selon la structure du JSON

    # Afficher les 5 premières lignes du DataFrame
    print(df.head())

    # Visualiser les données
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x="date_heure_tu", y="valeur", hue="nom_commune")
    plt.title("Concentration du polluant 24 en mars 2024")
    plt.xlabel("Date")
    plt.ylabel("Concentration")
    plt.xticks(rotation=45)
    plt.show()
else:
    print(f"Erreur {response.status_code}: {response.text}")
```

```python
import requests
import pandas as pd

# Fonction pour récupérer les données avec pagination
def fetch_data(url, params):
    data = []
    while url:
        response = requests.get(url, params=params)
        if response.status_code == 200:
            json_data = response.json()
            data.extend(json_data['results'])
            url = json_data.get('next')  # Récupérer l'URL de la page suivante
            params = None  # Supprimer les paramètres après la première requête
        else:
            print(f"Erreur {response.status_code}: {response.text}")
            url = None
    return data

# URL et paramètres de la requête
url = "https://data.airpl.org/api/v1/mesure/journaliere/"
params = {
    "code_configuration_de_mesure__code_point_de_prelevement__code_polluant": 3,
    "date_heure_tu__range": "2024-3-1,2024-3-31",
    "code_configuration_de_mesure__code_point_de_prelevement__code_station__code_commune__code_departement__in": "44,49,53,72,85",
    "export": "json"
}

# Récupérer toutes les données
data = fetch_data(url, params)

# Convertir en DataFrame
dfNO2 = pd.DataFrame(data)

# Sauvegarder le DataFrame pour utilisation ultérieure
dfNO2.to_pickle('NO2.pkl')

# Afficher les types de chaque colonne
print(dfNO2.dtypes)

# Afficher les premières lignes du DataFrame
print(dfNO2.head())
```
